To visualize the change in the number of total rail road kms I decided to make a line plot and the I just adjusted the plot size to analyze it easier.

```python
df.loc[['Year 1995','Year 1996','Year 1997','Year 1998','Year 1999',
        'Year 2000','Year 2001','Year 2002','Year 2003','Year 2004',
        'Year 2005','Year 2006','Year 2007','Year 2008','Year 2009',
        'Year 2010','Year 2011','Year 2012','Year 2013','Year 2014',
        'Year 2015','Year 2016','Year 2017','Year 2018','Year 2019'],
       'Total'].plot(kind='line', figsize=(30,7))
```

As we can see in the table above, the number of rail road kms has decreased almost half since 1995. Between 2005 and 2010 we can see a raise in this number but then it decreased, in 2015 it increased again but in 2016 it went down again.
