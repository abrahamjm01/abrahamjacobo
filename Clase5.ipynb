```python
path="/content/Canada.xlsx"
```

```python
import pandas as pd
```

```python
df=pd.read_excel(path,
                 sheet_name='Canada by Citizenship',
                 skiprows=range(20),
                 skipfooter=2
)
df
```

```python
df.columns
```

```python
to_be_droped=['Type', 'Coverage', 'AREA', 'REG','DEV','DevName']
```

```python
type(to_be_droped)
```

```python
df.drop(to_be_droped,
       axis=1,
       inplace=True)
df.head(2)
```

```python
new_names={'OdName':'Country',
           'AreaName':'Continent',
           'RegName':'Region'
           }
```

```python
df.rename(columns=new_names,
          inplace=True)
df.head(2)
```

```python
df.iloc[:,3:].sum(axis=1)
```

```python
df['Total']=df.iloc[:,3:].sum(axis=1)
```
