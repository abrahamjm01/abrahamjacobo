```python
path= 'https://raw.githubusercontent.com/tec03/Datasets/main/datasets/reg_data.csv'
```

```python
import pandas as pd
```

```python
df=pd.read_csv(path)
df.head(2)
```

```python
ndf=df.copy()
ndf.dropna(inplace=True)
```

```python
ndf
```

```python
#Salary of a person with age 33
#Salary= b0+b1*Age
```

```python
favs=['Salary', 'Age']
ndf=df[favs]
ndf.dropna(inplace=True)
ndf
```

```python
import statsmodels.api as sm
from sklearn import linear_model as LM
```

```python
#Salary= b0+b1*Age

X= pd.DataFrame(ndf.Age)
y= pd.DataFrame(ndf.Salary)
```

```python
type(X)
```

```python
lm=LM.LinearRegression()
model_lm=lm.fit(X, y)
model_lm
```

```python
print('Model intercept is', model_lm.intercept_)
print('Model coefficientes are', model_lm.coef_)
model_lm.score(X,y)
```
