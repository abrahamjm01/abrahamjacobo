```python
import statsmodels.api as sm
import pandas as pd
```

```python
path='/content/drive/MyDrive/Classroom/antropometria-dataset.csv'
```

```python
df= pd.read_csv(path)
df
```

```python
df
```

```python
ndf=df[['cadera','peso','entidad','sexo','diastol','sistol','edad','cintura']]
ndf.dropna(inplace=True)
ndf
```

```python
modelA=sm.OLS.from_formula('cadera~peso',data=ndf)
resultA=modelA.fit()
resultA.summary()
```

```python
a=resultA.resid
b=resultA.fittedvalues
```

```python
a
```

```python
from numpy.lib.function_base import median
median(b)
```

```python
modelB=sm.OLS.from_formula('entidad~sexo',data=ndf)
resultB=modelB.fit()
resultB.summary()
```

```python
modelC=sm.OLS.from_formula('diastol~sistol',data=ndf)
resultC=modelC.fit()
resultC.summary()
```

```python
modelD=sm.OLS.from_formula('diastol~sistol+edad+peso+cintura',data=ndf)
resultD=modelD.fit()
resultD.summary()
```
