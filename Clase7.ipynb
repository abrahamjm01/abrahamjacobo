```python
import statsmodels.api as sm
import pandas as pd
```

```python
path='https://raw.githubusercontent.com/tec03/Datasets/main/datasets/reg_data.csv'
```

```python
df= pd.read_csv(path)
df
```

```python
ndf=df[['Salary','Age','Gender','Company','JoinYear']]
ndf.dropna(inplace=True)
ndf
```

```python
modelA=sm.OLS.from_formula('Salary~Age',data=ndf)
resultA=modelA.fit()
resultA.summary()
```

```python
print('Salary of a person with age 33 is', 638.9240+41.5110*33 )
```

```python
ndf
```

```python
resultA.resid
resultA.fittedvalues
```

```python
modelB=sm.OLS.from_formula('Salary~Age+Gender',data=ndf)
resultB=modelB.fit()
resultB.summary()
```
